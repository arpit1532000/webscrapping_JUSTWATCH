## **Scrapping Movies Data**

````python
# Specifying the URL from which movies related data will be fetched
url='https://www.justwatch.com/in/movies?release_year_from=2000'
# Sending an HTTP GET request to the URL
page=requests.get(url)
# Parsing the HTML content using BeautifulSoup with the 'html.parser'
soup=BeautifulSoup(page.text,'html.parser')
# Printing the prettified HTML content
# print(soup.prettify())
````

## **Fetching Movie URL's**

````python
# Write Your Code here
base_url = "https://www.justwatch.com" # base url defined
movies_link_list = [] # creating a list

Movies_url = soup.find_all("a", class_="title-list-grid__item--link")[:n]
for i in Movies_url: # extracting url
    relative_url = i.get("href")
    http_link = base_url + relative_url
    movies_link_list.append(http_link) # appending urls
print(movies_link_list)
````

## **Now Creating Movies DataFrame**

````python
import requests
from time import sleep
from requests.exceptions import RequestException, HTTPError, Timeout
from bs4 import BeautifulSoup
import pandas as pd

def fetch_with_retries(url, retries=3, backoff_factor=0.3):
    attempt = 0
    while attempt < retries:
        try:
            response = requests.get(url)
            response.raise_for_status()
            return response
        except (HTTPError, Timeout, RequestException) as e:
            attempt += 1
            sleep_time = backoff_factor * (2 ** (attempt - 1))
            print(f"Attempt {attempt} failed: {e}. Retrying in {sleep_time} seconds...")
            sleep(sleep_time)
    raise RequestException(f"Failed to fetch {url} after {retries} retries")

movie_info_full_data = []
for l in movies_link_list:
  info_dict = {}
  url = l
  try:
    page = fetch_with_retries(url)
    soup = BeautifulSoup(page.text, 'html.parser')

    info_dict['Link'] = l
    info_dict['Title'] = soup.find_all('h1')[0].text.strip()
    info_dict['Release_year'] = eval(soup.find_all('span',class_='text-muted')[0].text.strip())

    try:
      streaming_platform = soup.find_all('div', class_ = 'buybox-row stream')[0].find_all('picture', class_ = 'picture-wrapper')[0].find_all('img')[0]['alt']
    except:
      streaming_platform = None
    info_dict['Streaming_platform'] = streaming_platform

    for i in soup.find_all('div', attrs={'class':'detail-infos'}):
      if i.find_all('h3', attrs={'class':'detail-infos__subheading'})[0].text == 'Genres':
        info_dict['Genres']=i.find_all('div', attrs = {'class':'detail-infos__value'})[0].text.strip()
      elif i.find_all('h3', attrs={'class':'detail-infos__subheading'})[0].text == 'Rating':
        info_dict['Rating']=i.find_all('div', attrs = {'class':'detail-infos__value'})[0].text.strip().split(' ')[0]
      elif i.find_all('h3', attrs={'class':'detail-infos__subheading'})[0].text == 'Runtime':
        info_dict['Runtime']=i.find_all('div', attrs = {'class':'detail-infos__value'})[0].text.strip()
      elif i.find_all('h3', attrs={'class':'detail-infos__subheading'})[0].text == 'Age rating':
        info_dict['Age_rating']=i.find_all('div', attrs = {'class':'detail-infos__value'})[0].text.strip()
      elif i.find_all('h3', attrs = {'class':'detail-infos__subheading'})[0].text.strip() =='Production country':
        info_dict['Production_country'] = i.find_all('div', attrs = {'class':'detail-infos__value'})[0].text.strip()

    movie_info_full_data.append(info_dict)
    print('successful',l)
  except RequestException as e:
    print('error',l,e)
    continue

  sleep(2)

movie_df = pd.DataFrame(movie_info_full_data)
````

## **Creating TV Show DataFrame**

````python
# Write Your Code here
import requests
from time import sleep
from requests.exceptions import RequestException, HTTPError, Timeout
from bs4 import BeautifulSoup
import pandas as pd

def fetch_with_retries(url, retries=3, backoff_factor=0.3):
    attempt = 0
    while attempt < retries:
        try:
            response = requests.get(url)
            response.raise_for_status()
            return response
        except (HTTPError, Timeout, RequestException) as e:
            attempt += 1
            sleep_time = backoff_factor * (2 ** (attempt - 1))
            print(f"Attempt {attempt} failed: {e}. Retrying in {sleep_time} seconds...")
            sleep(sleep_time)
    raise RequestException(f"Failed to fetch {url} after {retries} retries")

tv_info_full_data = []
for t in tv_show_link_list:
  infotv_dict = {}
  url = t
  try:
    page = fetch_with_retries(url)
    soup = BeautifulSoup(page.text, 'html.parser')

    infotv_dict['Link'] = t
    infotv_dict['Title'] = soup.find_all('div',class_ = 'title-block')[0].find_all('h1')[0].text.strip()
    infotv_dict['Release_year'] = eval(soup.find_all('div',class_ = 'title-block')[0].find_all('span', class_ = 'text-muted')[0].text.strip())

    try:
      streaming_platform = soup.find_all('div', class_ = 'buybox-row stream')[0].find_all('picture', class_ = 'picture-wrapper')[0].find_all('img')[0]['alt']
    except:
      streaming_platform = None
    infotv_dict['Streaming_platform'] = streaming_platform

    for i in soup.find_all('div', attrs={'class':'detail-infos'}):
      if i.find_all('h3', class_ = 'detail-infos__subheading')[0].text.strip() == 'Genres':
        infotv_dict['Genres'] = i.find_all('div', class_ = 'detail-infos__value')[0].text.strip()
      elif i.find_all('h3', class_ = 'detail-infos__subheading')[0].text.strip() == 'Rating':
        infotv_dict['Rating'] = i.find_all('div', class_ = 'detail-infos__value')[0].text.strip().split(' ')[0]
      elif i.find_all('h3', class_ = 'detail-infos__subheading')[0].text.strip() == 'Age rating':
        infotv_dict['Age_rating'] = i.find_all('div', class_='detail-infos__value')[0].text
      elif i.find_all('h3', class_ = 'detail-infos__subheading')[0].text.strip() == 'Runtime':
        infotv_dict['Runtime'] = i.find_all('div', class_ = 'detail-infos__value')[0].text
      elif i.find_all('h3', class_ = 'detail-infos__subheading')[0].text.strip() == 'Production country':
        infotv_dict['Production country'] = i.find_all('div', class_ = 'detail-infos__value')[0].text.strip()

    tv_info_full_data.append(infotv_dict)
    print('successful',t)
  except RequestException as e:
    print('error',t,e)
    continue

  sleep(2)

tv_df = pd.DataFrame(tv_info_full_data)
````
